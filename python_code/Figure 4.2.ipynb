```python
x = np.arange(0,agents+1)
prob_eq = np.load(path_data+'prob_eq.npy')
ex_eq_v = np.load(path_data+'ex_eq_v.npy')
ex_eq_good = np.load(path_data+'ex_good.npy')
ex_eq_bad = np.load(path_data+'ex_bad.npy')
ex_eq_other = np.load(path_data+'ex_other.npy')

fig = plt.figure(figsize=(12,6),tight_layout=True,dpi=150)
gs = fig.add_gridspec(4, 1)
ax1 = fig.add_subplot(gs[:2, 0])
ax2 = fig.add_subplot(gs[2:, 0])

for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(0.25)

ax1.set_title('Panel 1: Expected number of total equilibria as a function of R', fontsize= 'medium')
ax1.set_xticks(np.arange(0,agents+10,10))
ax1.set_xlim([0, agents])
ax1.set_yticks(np.arange(0,3.5, 0.25))
ax1.set_ylim([0.75,2.5])
ax1.plot(x, ex_eq_good+ex_eq_bad+ex_eq_other, color='k', linestyle='solid',linewidth=0.5,  
         label='Expected number of equilibria')
ax1.legend(loc='best', fontsize=11, frameon=False)

ax2.set_title('Panel 2: Expected number of good and bad equilibria as a function of R', fontsize= 'medium')
ax2.set_xlabel('Enforcement resources (R)', fontsize='large')
ax2.set_xticks(np.arange(0,agents+10,10))
ax2.set_xlim([0, agents])
ax2.set_yticks(np.arange(0,1.5, 0.25))
ax2.set_ylim([0,1.25])
ax2.plot(x, ex_eq_good, color='k', linestyle='solid',linewidth=0.75, label='# of equilibria $v \leq 30$')
ax2.plot(x, ex_eq_bad, color='k', linestyle='dashed',linewidth=0.75, label='# of equilibria $v \geq 70$')
ax2.legend(loc='best', fontsize=11, frameon=False)

#plt.savefig(path_figs+'new fig', facecolor='w');
```
