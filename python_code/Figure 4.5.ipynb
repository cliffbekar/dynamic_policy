## Figure 4.5: The cliff of expected violations.

```python
agents = 100

ex_v_d = np.load(path_data+'ex_v_d-model.npy')   
ex_v_r = np.load(path_data+'ex_v_r-model.npy')
ex_v_p = np.load(path_data+'ex_v_p-model.npy')

x = np.arange(0,agents+1)

fig = plt.figure(figsize=(12,4),tight_layout=True,dpi=250)
gs = fig.add_gridspec(3, 1)
ax1 = fig.add_subplot(gs[:3, 0])

for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(0.25)

plt.xlabel('Enforcement resources (R)', fontsize= 'medium')
ax1.set(ylabel='Number of violators')
ax1.yaxis.label.set_size(11)
ax1.xaxis.label.set_size(11)
ax1.set_xticks(np.arange(0,agents+1,10))
ax1.set_yticks(np.arange(0, agents+2,10))
ax1.set_xlim([0, agents+1])
ax1.set_ylim([0, agents+2])
ax1.plot(x, ex_v_d, color='k', linestyle='solid',linewidth=0.75, label=r'$E(v|R)$, D-model')
ax1.plot(x, ex_v_r, color='gray', linestyle='dashed',linewidth=0.75, label=r'$E(v|R)$, R-model')
ax1.plot(x, ex_v_p, color='gray', linestyle='dotted',linewidth=0.75, label=r'$E(v|R)$, P-model')
ax1.legend(loc='best', fontsize=10, frameon=False)

#plt.savefig();
```
